```python
from sklearn.cluster import KMeans
import pandas as pd
```

```python
datapath ="C:/Users/garim/Documents/Capstone/Full data/Normalized/mirna_BRCA_s01s11.csv"
mirna = pd.read_csv(datapath)
```

```python
tmirna = pd.DataFrame(mirna.transpose())
```

```python
tmirna.columns = tmirna.iloc[0,:]
```

```python
tmirna = tmirna.drop('Unnamed: 0')
```

```python
tmirna.head()
```

```python
tmirna_sample = tmirna.iloc[:,-1]
```

```python
tmirna = tmirna.drop('sampleC', axis = 1)
```

```python
sample = pd.DataFrame(list(tmirna_sample))
```

```python
kmeans = KMeans(n_clusters = 2)
kmeans_tmirna = kmeans.fit(tmirna)
labels = kmeans_tmirna.predict(tmirna)
centroids = kmeans_tmirna.cluster_centers_
```

```python
labels = pd.DataFrame(list(labels))
```

```python
labels.head()
```

```python
sample.head()
```

```python
classes = pd.concat([labels, sample], axis=1)
classes.columns = ['cluster','sample']
```

```python
classes.head()
```

```python
classes.groupby(['sample'])['cluster'].sum()
```
