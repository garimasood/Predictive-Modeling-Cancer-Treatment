```python
mirna = mirna.drop('sampleC')
```

```python
kmeans = KMeans(n_clusters = 20)
kmeans_mirna = kmeans.fit(mirna)
labels = kmeans_mirna.predict(mirna)
centroids = kmeans_mirna.cluster_centers_
```

```python
labels = pd.DataFrame(list(labels))
```

```python
labels = labels.set_index(mirna.index)
```

```python
clustered_dat = pd.concat([mirna,labels], axis =1,join_axes=[mirna.index])
```
